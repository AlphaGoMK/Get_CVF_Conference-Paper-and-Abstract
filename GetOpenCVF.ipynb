```python
#encoding:utf-8 
from bs4 import BeautifulSoup
from urllib.request import urlopen
import codecs
```

```python
pageurl="http://openaccess.thecvf.com/CVPR2018.py"
soup = BeautifulSoup(urlopen(pageurl).read().decode('utf-8'),'lxml')
content=soup.find(id='content')
paperNameList=content.find_all('dt')
```

### Print `paper_name: abstract` into `{conference_name}.txt`

```python
filename=pageurl.split('/')[-1].split('.')[0]+'.txt'

for item in paperNameList:
    atag=item.find_all('a')[0]
    print(atag.string)
    
    url='http://openaccess.thecvf.com/'+atag['href']
    subcontent=BeautifulSoup(urlopen(url).read().decode('utf-8'),'lxml')
    abstract_tag=subcontent.find(id='abstract')
    
    title=atag.string
    abstract=abstract_tag.string.strip()
    # Support Franch
    with codecs.open(filename,"a", "utf-8") as f:
        f.write(atag.string+":")
        f.write(abstract_tag.string+"\n")
```
