```python
import csv
import os
import requests
import subprocess
csvFile = open("youtube_boundingboxes_detection_train.csv", "r")
reader = csv.reader(csvFile)
```

```python
urlpost=[]
anno=[]
filenamelist={}

# my_proxies={"http":"http://127.0.0.1:1080","https":"https://127.0.0.1:1080"}
# subprocess.call("youtube-dl http://youtu.be/AAB6lO-XiKE --proxy socks5://127.0.0.1:1080", shell=True)
# os.system("youtube-dl"+" http://youtu.be/AAB6lO-XiKE")

print('ok')
# r=requests.get("http://youtu.be/AAB6lO-XiKE", proxies=my_proxies)
# print(r)

# with open("a.mp4","wb") as f:
#     f.write(r.content)
```

```python
cnt=0
for item in reader:
    
    label=item[3]
    if not os.path.exists('train/video/'+label):
        os.mkdir('train/video/'+label)
        
    if not os.path.exists('train/anno/'+label):
        os.mkdir('train/anno/'+label)
        
    if item[0]+'_'+str(item[2])+'_'+str(item[4]) not in urlpost:
        cnt+=1
        
        if cnt>=1000:
            break
            
        url='http://youtu.be/'+item[0] 
        videofilename='train/video/'+label+'/'+item[0]+'_'+str(item[2])+'_'+str(item[4])+'.mp4'           
        cmd="youtube-dl -o "+videofilename+" "+url+" --proxy socks5://127.0.0.1:1080"
        print(str(cnt)+": "+cmd)
        subprocess.call(cmd, shell=True)
        urlpost.append(item[0]+'_'+str(item[2])+'_'+str(item[4]))
            
    annofilename='train/anno/'+label+'/'+item[0]+'_'+str(item[2])+'_'+str(item[4])+'.txt'
    with open(annofilename, "a+") as f:
        newlist=item[1:]
        s=""
        for idx,it in enumerate(newlist):
            s+=str(it)
            if idx!=len(newlist)-1:
                s+=", "
            else:
                s+='\n'

        print(s)
        f.write(s)
        

        
```
